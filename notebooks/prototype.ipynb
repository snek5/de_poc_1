```python
# Necessary imports
import pandas as pd
import fastf1
import duckdb
```

```python
# Creating a connection to duckdb
con = duckdb.connect('f1analysis')
```

```python
# Init data
session = fastf1.get_session(2025, 1, 'R')
session.load()
```

```python
# Store raw laps data
laps_df = session.laps
con.execute("CREATE OR REPLACE TABLE bronze_laps AS SELECT * FROM laps_df")
```

```python
# Verify
print(con.execute("SELECT * FROM bronze_laps WHERE Driver = 'VER'").df())
```
